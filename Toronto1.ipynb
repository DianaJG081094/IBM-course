# Task 1

```python
import pandas as pd
#Obtein data from html
table_codes = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df = table_codes[0]

#Three columns:PostalCode,Borough,Neighborhood
df.columns = ['PostalCode','Borough','Neighborhood']

#Ignore cells with a borough that is Not Asiigned
df = df[df['Borough'] != 'Not assigned']

# Not assigned neighborhood case: neighborhood will be the same boroutgh
df.Neighborhood.replace('Not assigned',df.Borough,inplace=True)

#Listed neighborhoods
df = df.groupby(['PostalCode','Borough'])['Neighborhood'].apply(lambda x: ','.join(x)).reset_index()

df
```

#### Analize each Borough

```python
borough_study = pd.get_dummies(venues_df[['Category']], prefix="", prefix_sep="")

borough_study['Borough'] = venues_df['Borough']

fixed_columns = [borough_study.columns[-1]] + list(borough_study.columns[:-1])
borough_study = borough_study[fixed_columns]

borough_study.head()
```

#### Analize Borough by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = borough_study.groupby('Borough').mean().reset_index()
toronto_grouped
```

#### Look for the top 5 most common venues at each Borough

```python
num_top_venues = 10

for hood in toronto_grouped['Borough']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Borough'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#To DataFrame
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Borough']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
borough_venues_sorted = pd.DataFrame(columns=columns)
borough_venues_sorted['Borough'] = toronto_grouped['Borough']

for ind in np.arange(toronto_grouped.shape[0]):
    borough_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

borough_venues_sorted.head()
```

### Clustering

```python
#Number of clusters
kclusters = 2

toronto_grouped_clustering = toronto_grouped.drop('Borough',1)

#K means
kmeans = KMeans(n_clusters=kclusters,random_state=0).fit(toronto_grouped_clustering)

kmeans.labels_[0:10]
```

```python
# add clustering labels
borough_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
borough_venues_sorted.head()
```

```python
#borough_venues_sorted
toronto_f = toronto_df
toronto_f = toronto_f.join(borough_venues_sorted.set_index('Borough'),on='Borough')
toronto_f.head()
```

```python
#map clusters
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_f['Latitude'], toronto_f['Longitude'], toronto_f['Neighborhood'], toronto_f['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
